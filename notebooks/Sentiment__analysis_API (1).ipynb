```python
from getpass import getpass
GITHUB_TOKEN = getpass("Paste your GitHub PAT (input is hidden): ")
print("Token stored in memory (not printed).")
```

```python
from openai import OpenAI
client = OpenAI(
base_url="https://models.github.ai/inference",
api_key=GITHUB_TOKEN
)
print("Client created successfully.")
```

```python
response = client.chat.completions.create(
    model="meta/Llama-4-Scout-17B-16E-Instruct",
    temperature=0,
    top_p=0.74,
    max_tokens=100,
    messages=[
        {
            "role": "user",
            "content": "tell me if the sentence is one of:  Positive / Negative / Neutral + a short explanation for these and put them in a table, 1. 'I dont care about the book', 2. 'I like this painting', 3.'Why did we add this?', 4.'I miss home'",
        }
    ]
)

print("Model response:\n")
print(response.choices[0].message.content)
```
